#### Import NumPy as np

```python
import numpy as np
# I have imported numpy and named it np.
```

#### Create an array of 20 zeros 

```python
np.zeros(20)
# I made a list of 20 zeros into an array.
```

#### Create an array of 10 ones

```python
np.ones(10)
# I made an array out of a list of ten ones.
```

#### Create an array of 10 fives

```python
np.ones(10)*5
# i made an array out of a list of ten ones and multiplied all the values by five
```

#### Create an array of the integers from 10 to 50

```python
np.arange(10,50)
# I made an array of integers with a range of 10 to 50.
```

#### Create an array of all the even integers from 10 to 50

```python
np.arange(10,50,2)
# I made an array of integers divisible by 2 with a range of 10 to 50.
```

#### Create a 3x3 matrix with values ranging from 0 to 8

```python
matrix=np.arange(0,9)
#formed an array with values ranging from 0 to 8
matrix.reshape(3,3)
# reformatted the matrix into three lists, each with three values
```

#### Create a 3x3 identity matrix

```python
np.eye(3)
#I made a matrix with three lists of three values each, with the main diagonal value.
```

#### Use NumPy to generate a random number between 0 and 1

```python
np.random.rand(1)
# I made an array with a random integer between 0 and 1 in it.
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

```python
np.random.rand(25)
#I made a 25-number array with values ranging from 0 to 1.
```

#### Create the following matrix:

```python
np.arange(0.01,1.01,.01).reshape(10,10)
# I made a matrix with increments of.01 between 0.01 and 1.01
# I reformatted the matrix into ten lists of ten values each.
```

#### Create an array of 20 linearly spaced points between 0 and 1:

```python
np.random.rand(20)
# I made a 20-number array with values ranging from 0 to 1.
```
