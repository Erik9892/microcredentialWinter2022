while loops

While loops are used to search through data; while a certain condition is True.

the diff btwn while and for lopps:

while loop  first checks if the test is valid or false. if its valid it will continiue to run and check again til its invalid (false)

for loop runs  the fisrt  literation or   then checks  if theres nothin else to tun. 

```python
number = 1

while number <= 5:
  print(number)
  number += 1
#The first line we start our counting from the number = 1.
#The while loop is set to keep running while a certain condition is true. In this case, while number is <= 5.
#The code inside the loop prints the current numbe (print(number)) then increases the number by 1 (+= 1). += basically means: number = number + 1.
#Python repeats the loop until the condition is no longer true.
```

WHILE LOOPS AND USER INPUT

```python
# A while loop can be broken by a users input.
prompt = 'Input message... I will repreat untill you type quit...: '
message = ''

while message != 'quit':
  message = input(prompt)
  print(message)
```

FLAGS WITH WHILE LOOPS

Sometimes it makes more sense to create a flag to signal to Python to break the while loop.

```python
#Programs will run while the flag is set to True.
prompt = 'Input message... I will repreat untill you type quit...: '
message = ''


active = True

while active:
  message = input(prompt)
  
  if message == 'quit':
    active = False
  else:
    print(message)
```
