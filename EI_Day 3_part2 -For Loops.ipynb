# For Loops

Looping allows you to essential ‘automate’ a task.  
allowing  you to perform the same task over and over on a list of data points.

A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).


```python
#A for loop can be used to loop through and print the letters in a string. 
# we start a a for loop with for _ in  "" and you end with a colon
for x in 'CUNY':
  print(x)
```

```python
for c in 'Tv shows':
  print (c)
```

```python
# For loops can also loop through a list of items
TV_Shows = ['game of thrones', 'star trek', 'stranger things']
```

```python
for show in TV_Shows:
  print(show)
  
# We start by defining a list.
# We then define a for loop – this tells Python to pull a name from the list tvshows  and store the variable in show.
# We then tell Python to print the tvshows that were just stored in show.
#Python then repeats items 2 and 3 until all of the tvshows are printed.
#Another way to think about it is: “For every show in the list of tvshows, print the tv show”
```

Adding on to our previous for loop

```python
for show in TV_Shows:
  print(show.upper() + ' is one of my fav TV shows')
```

```python
#The range() function can also be used to make a list [].
numbers = list(range(1,11))
print(numbers)
```

```python
#The range() function can also take another argument.
numbers = list(range(0,11,2)) 
print(numbers)
# it wiil run the list but it  will count by 2 (step)
```

```python
# it will run the list by a step of 3 
for x in range(1,21):
  print(x+3)
```

SLICING YOUR LIST

You can slice through your list to print out only certain groups of elements.

```python
tvshows = ['game of thrones', 'star trek', 'stranger things', 'dc', 'true blood' ]
```

```python
print(tvshows[0:3])
```

```python
print(tvshows[2:4])
```

```python
print(tvshows[:3])
```

```python
print(tvshows[2:4])
```

```python
print(tvshows[2:])
```

```python
print(tvshows[-2:])
```

You can even loop through your slice with a for loop.

```python
tvshows = ['game of thrones', 'star trek', 'stranger things', 'dc', 'true blood']
print("some cool tv shows")
for coolshows in tvshows[:2]:
  print(coolshows.title())
```
