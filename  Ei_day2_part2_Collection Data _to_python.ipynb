## List
 
1.   A list in Python is a collection of items in a specific order.
2.  Lists can consist of letters, digits, names of people, etc.  Anything can be placed in a list.
3.In Python, lists are indicated by square brackets ( [] ) – elements within a list are separated by commas.








```python
movies = ['Star Wars', 'Star Trek', 'Lord of the Rings', 'Avengers', 'The Hobbit', "Fast and Furious"]
print(movies)
```

```python
# findind the number of characters or FINDING THE LENGTH OF LIST
len(movies)
```

```python
#INDEX POSITIONS START AT 0, NOT 1
#Python considers the first item within a list to be at position 0, NOT 1.
movies = ['Star Wars', 'Star Trek', 'Lord of the Rings', 'Avengers', 'The Hobbit', "Fast and Furious"]
print(movies[0])
print(movies[1])
print(movies[2])
print(movies[3])
```

```python
# last movies on list
movies[-1]
```

```python
# another way to find the last movies on the list
print(movies[-1])
```

```python
# Methods can also be used on strings,
print(movies[0].title())
print(movies[0].upper())
```

LIST VARIABLES

You can use values from a list like other variables.

```python
movies = ['Star Wars', 'Star Trek', 'Lord of the Rings', 'Avengers', 'The Hobbit', "Fast and Furious"]
# In the previous line I am creating a list of movies.

message = 'My favourite movie is ' + movies[0].upper() + '!!!'
'''
In the previous line I am first creating a variable. Then I am assigning a couple of things.
first, I am extracting a movie from my list by indexing through the list.
then, I am applying the built-in method (or function) '.upper()' to the extracted movie from 
the list. The '.upper()' method simply capitalises every letter -  e.g. word --> WORD.
Lastly, I am concatinating a string to my extracted movie from the list and then I am 
concatinating another string at the end.

'''

print(message)
# In this line I am printing the content of my varaible 'message' to the screen
```

MODIFYING LISTS

Python allows you to make modifications to your lists – adding and
removing elements from your list. 

APPENDING MORE DATA TO YOUR LIST

Often times, you’ll need to expand your list and add more data. Python allows this to be done using a simple method called append().

```python
# adding an appending to my list
movies.append('i love this movies')
print(movies)
```

```python
#Python also allows you to insert elements into your list.
#INSERTING the green mile in first position
movies.insert(0,'green mile')
print(movies)
```

```python
movies2 = ['Speed', 'Archive']
```

```python
movies.append(movies2)
```

```python
movies
```

REMOVING ELEMENTS FROM A LIST

If you know which position the element you want to remove is in, you can use the del statement to remove that element.

```python
# del element from list 
print(movies)
del movies[0]
print (movies)
```

```python
#another way of del
movies.remove(movies2)
```

```python
print (movies)
```

SORTING A LIST

Lists can be sorted using the sort() method.

```python
print(movies)
movies.sort()
print(movies)
```

```python
#Lists can also be sorted in reverse
print(movies)
movies.sort(reverse=True)
print(movies)
```

```python
# Convert a integer number into Binary, Octal and Hexadecimal

num = int(input("Enter number: "))

print(f'Binary value is {bin(num)}')
print(f'Octal value is {oct(num)}')
print(f'Hexadecimal value is {hex(num)}')

print(f'Binary value is {bin(num)}, Octal value is {oct(num)}, Hexadecimal value is {hex(num)}')
```

# DICTIONARIES

Dictionaries in Python allow you to connect different pieces of information
together.

The data stored in dictionaries can be looped through similar to what we did
in lists.

Dictionaries allow you to store two kinds of information that can then be
matched up.

```python
#example
#The dictionary patient_1 stores the patients first name and last name. 
# The two print statements display the information stored.
patient1_= {'first_name': 'john', 'last_name': 'kimble' }
```

```python
print(patient1_['first_name'])
print(patient1_['last_name'])
```

key values

A dictionary in Python is a collection of key-value pairs.

Each key is connected to a value. You can use a key to
access the value associated with that key.

A key’s value can be a number, list, string, or even another
dictionary.



```python
# Python Dictionaries are wrapped in braces, {}. With their key-value pairs inside the braces.
# A key-value pair is a set of values associated with each other.
# Every key is connected to its value by a colon, and individual
# key-value pairs are separated by commas.
patient_1= {'first_name': 'john', 'last_name': 'kimble' }
```

ACCESSING VALUES

To get the value associated with a key: give the name of the dictionary and placethe key inside a set of square brackets.

You can have an unlimited # of key-value pairs in a dictionary.

```python
patient1_= {'first_name': 'john', 'last_name': 'kimble' }
print(patient1_['first_name'])
print(patient1_['last_name'])
```

```python
# creating a list To get the value associated with a key
fruit_1= {'fruit_name': 'Granny Smith', 'fruit_type': 'apple' }
```

```python
print(fruit_1['fruit_name'])
print(fruit_1['fruit_type'])
```

ACCESSING DICTIONARIES

```python
patient_name = patient_1['first_name']
print('hello there,' + patient_name)
```

```python
fruit_name = fruit_1['fruit_name']
print(' i love,' " " + ""  + fruit_name)
```

Adding NEW KEY-VALUE PAIRS

Data is always changing – or dynamic.
It makes sense that you are able to modify your dictionaries.

To add a new key-value pair, you would give the name of the dictionary followed by the new

key in square brackets along with the new value.

```python
# example
patient_1['mrd'] = '000000000'
patient_1['lastexam'] = 'echo'
```

```python
patient_1
```

STARTING WITH AN EMPTY DICTIONARY

```python
# To start filling an empty dictionary, first define a dictionary
patient_2={}
# Then, add each key-value pair on its own line.
patient_2['first_name'] = 'jean'
patient_2['last_name'] = 'picard'
```

```python
print(patient_2)
```

MODIFYING VALUES

To modify a value,, give the name of the dictionary with the key in []’s and then the new value you want associated with that key.

```python
#edditing  value
patient_2['first_name'] = 'jean-luc'
```

```python
print(patient_2)
```

REMOVING KEY-VALUE PAIRS

Data stored within the key-value pair can be easily removed.

You can use the del statement to completely remove a key-value
pair – this is permanent.

```python
del patient_2['first_name']
print(patient_2)
```
