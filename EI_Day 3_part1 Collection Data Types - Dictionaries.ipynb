# Dictionaries

Dictionaries in Python allow you to connect different pieces of information together.

The data stored in dictionaries can be looped through similar to what we did in lists.

Dictionaries allow you to store two kinds of information that can then be matched up.

```python
#example
#The dictionary patient_1 stores the patients first name and last name. 
# The two print statements display the information stored.
patient1_= {'first_name': 'john', 'last_name': 'kimble' }
```

key values

A dictionary in Python is a collection of key-value pairs.

Each key is connected to a value. You can use a key to access the value associated with that key.

A key’s value can be a number, list, string, or even another dictionary.

```python
# Python Dictionaries are wrapped in braces, {}. With their key-value pairs inside the braces.
# A key-value pair is a set of values associated with each other.
# Every key is connected to its value by a colon, and individual
# key-value pairs are separated by commas.
patient_1= {'first_name': 'john', 'last_name': 'kimble' }
```

ACCESSING VALUES

To get the value associated with a key: give the name of the dictionary and placethe key inside a set of square brackets.

You can have an unlimited # of key-value pairs in a dictionary.

```python
# creating a list To get the value associated with a key
patient1_= {'first_name': 'john', 'last_name': 'kimble' }
```

```python
print(patient1_['first_name'])
print(patient1_['last_name'])
```

ACCESSING DICTIONARIES 

```python
patient_name = patient_1['first_name']
print('hello there,' + patient_name)
```

Adding NEW KEY-VALUE PAIRS

Data is always changing – or dynamic. It makes sense that you are able to modify your dictionaries.

To add a new key-value pair, you would give the name of the dictionary followed by the new

key in square brackets along with the new value.

```python
# example
patient_1['mrd'] = '000000000'
patient_1['lastexam'] = 'echo'
```

To modify a value,, give the name of the dictionary with the key in []’s and then the new value you want associated with that key.

REMOVING KEY-VALUE PAIRS

Data stored within the key-value pair can be easily removed.

You can use the del statement to completely remove a key-value pair – this is permanent.

```python
del patient_2['Object']
```

Dictionaries can store lots of information about the object, easily and in diffrent formats:

We can then access elements of our dictionary.

```python
patient_3 = {'Name': 'Jack', 'Last Name': 'Sparrow', 'MRN': '001234567', 'DOB': '22231845'}
patient_3
```

```python
# trying out a diffrent format
patient_3 = {
    'Name': 'jack',
    'Last Name':'sparrow',
    'MRN': '001234567' ,
    'DOB': '22231845'
}
print(patient_3)
```

```python
# Using string formating 
print(f'The first name of the Patient is {patient_3["Name"]}')
print(f'The last name of the Patient is {patient_3["Last Name"]}')
print(f'The ID of the Patient is {patient_3["MRN"]}')
print(f'The DOB of the Patient is {patient_3["DOB"]}')
```

# Nesting with Dictionaries

```python
# Dictionary nested inside a dictionary nested in side a dictionary
d2 = {'key1':{'nestkey':{'subnestkey':'value'}}}
```

```python
d2.items()
```

```python
d2['key1']['nestkey']['subnestkey']
```

# Dictionary Methods

```python
# Create a typical dictionary
d = {'key1':1,'key2':2,'key3':3}
```

```python
# Method to return a list of all keys 
d.keys()
```

```python
# Method to grab all values
d.values()
```

```python
# Method to return tuples of all items  
d.items()
```
