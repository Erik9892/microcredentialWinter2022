# Slicing
We can use a : to perform slicing which grabs everything up to a designated point.

```python
# Grab everything past the first term all the way to the length of s which is len(s)
s[1:]
```

```python
# Note that there is no change to the original s
s
```

```python
# Grab everything UP TO the 3rd index
s[:3]
```

Note the above slicing. Here we're telling Python to grab everything from 0 up to 3. It doesn't include the 3rd index. You'll notice this a lot in Python, where statements and are usually in the context of "up to, but not including".

```python
#Everything
s[:]
```

```python
# We can also use negative indexing to go backwards.
# Last letter (one index behind 0 so it loops back around)
s[-1]
```

```python
# second to Last letter 
s[-2]
```

```python
# Grab everything but the second to last and last letter
s[:-2]
```

```python
# Grab everything but the , thrid, second, to last and last letter
s[:-3]
```

```python
# Grab everything but the fifth, fourth, thrid, second, to last and last letter
s[:-5]
```

# Strings Concatenation

It’s often useful to combine strings. Lets say you gretting someone and their name stored in different columns. When you’re generating a report, you might find it useful to combine both
of their values into one.

```python
#adding Variables  
greeting = "Hello"
name = "Mark"
#combaning Strings
print(greeting + name)
#adding space when combaning
print(greeting + " " + name)
```

```python
# adding a number to a string 
print(greeting + str(6))
```

```python
# how to find a stringn line 
print(greeting + '6')
```
